# Что такое Python и что нам нужно
***

Для запуска программы написанной на Python'e необходим интерпретатор.

Интерпретатор - программа, которая служит для выполнения кода.

Интерпретатор пошагово исполняет код исходя из общепринятых правил.

Программа на Python'e - это текст кода, сохраненный в формате ".py".

Чтобы исполнить написанный вами код необходимо вызвать интерпретатор и указать путь к файлу программы.

Пример: 

```sh
python имя_файла.py 
```

# Условные операторы и ~~Операторные скобки~~
Условные конструкции в Python существуют для того,  
чтобы разработчик мог задавать определенное поведение  
программы в зависимости от заданных ей условий.  

### Оператор if
Оператор if в Python подобен этому оператору в других языках.  
Утверждение с использованием оператора if содержит в себе логическое условие,  
в котором производится сравнение данных и по результату выбирается дальнейшее действие.  
Пример:
```python
if выражение:
    действие
```
Вслучае если **выражение** окажется истинным,  
интерпретатор выполнит команды из _вложенного блока_.
&nbsp;
    
### Вложенный блок == ~~операторные скобки~~
Python **не содержит** операторных скобок (begin..end в pascal или {..}в Си),  
вместо этого блоки выделяются отступами: **4 пробелами**,  
а вход в блок из операторов осуществляется двоеточием.

````python
# например
num = -1
if num > 0:
    print("Если ты это читаешь, значит num > 4")
    print("И эту строчку ты читаешь, потому что передомной тоже есть отступ")
print("На этой строчке отступов нет...")
print("Это значит, что в любом случае, ты это прочтешь.")
````

# Оператор else - иначе
В случае, когда необходимо выполнить действия при ложном  
выражении, описанном после оператора **if**  
можно использовать конструкцию **if .. else**.  
Например:
```python
if выражение:
    действие_1
else:
    действие_2 выполниться если выражение оказалось ложным
```
    **Примечание** Конструкция **else** является необязательной.
&nbsp;
&nbsp;
## Конструкция if..elif..else:
Ключевое слово **elif** добавляется в том случае, если необходимо  
проверить _несколько условий не связанных между собой_.  
"Несколько условий" означает, что указывать **elif**  
можно сколько угодно раз.
```python
if выражение:
    действия, если выражение равно True
elif выражение2:
    действия, если выражение2 равно True
elif выражение3:
    действия, если выражение3 равно True
else:
    действия, в противном случае
```


````python
if 2 > 3 < 4 and 1 or 0:
    print('yes')
````
