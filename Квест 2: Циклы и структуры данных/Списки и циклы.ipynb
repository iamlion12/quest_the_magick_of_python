# Списки

Списки Python схожи с *массивами* в других языках. В Python, пустой список может быть создан следующим образом:

```python
my_list = []
my_list = list()
```
Как вы видите, вы можете создать список при помощи квадратных скобок,  
или при помощи встроенного инструмента Python – list. Список состоит любых объектов, количество неограниченно.

**Операции со списками**
* **len**(список) - *длина списка  *
* **sorted**(список) - *сортировка списка  *
* объект **in** список - *проверка вхождения элемента в список  *
* список.**append**(объект) - *добавление элемента в конец * 
* список**.extend**(другой_список) - *расширение списка*
* список **+** список -* конкатенация(склеивание) списков*

**Доступ к элементам**  
Индекс - целое число. В Списках и Кортежах и в строках нумерация начинается с нуля.

Пример:
```python
список[индекс]
```

# Цикл while

While - один из самых универсальных циклов в Python, поэтому довольно медленный. Выполняет тело цикла до тех пор, пока условие цикла истинно.

```python
while выражение:
    # тело цикла во вложенном блоке
    действия
```

# Цикл **for**

Цикл for уже чуточку сложнее, чуть менее универсальный, но выполняется гораздо быстрее цикла while. Этот цикл проходится по любому итерируемому объекту (например строке или списку), и во время каждого прохода выполняет тело цикла.

```python
for имя переменной in объект:
    # тело цикла во вложенном блоке
    действия
```

# range()

Итак, как работает функция Python под названием range? Простыми словами, range() позволяет вам генерировать ряд чисел в рамках заданного диапазона. В зависимости от того, как много аргументов вы передаете функции, вы можете решить, где этот ряд чисел начнется и закончится, а также насколько велика разница будет между двумя числами. range принимает только целочисленные аргументы.
```python
for имя_переменной in range(начало,конец,шаг):
  действия
```

### Оператор in 

````python
# Проверить входит ли символ в строку:
print('1' in 'Во мне есть единица?')
````

````python
# объявление списка
my_list = ['Nif-Nif', 23, 1.342, None, True, ['Хей!', "Я список в списке!"]]
print(my_list)

# индексы
print("\nВывод элементов по индексу:")
print(my_list[1])
print(my_list[2] + 2)
print(my_list[4] == True)
````

````python
# проверить входит ли объект в список
print('Nif-Nif' in my_list)
````

````python
# Как изменить i-ый элемент?
my_list[0] = 'Начало'
print(my_list)
````

````python
# длинна списка
len(my_list)
````

````python
# добавления объекта в конец
my_list.append('Конец')
print(my_list)
````

````python
# "расширить" список
my_list.extend(my_list)
print(my_list)
````

````python
# сложение списков
print(['1', '2', '3'] + [127, 0, 0, 1])
````

````python
# список элементов от 0 до 10
list(range(10))
````

````python
# вывод на экран чисел от 0 до 10
for i in range(10):
    print(i)
````

````python
# вывод на экран первые 7 елементов списка
for i in range(7):
    print(my_list[i])
````

````python
# тоже самое с помощью цикла while
i = 0
while i < 7:
    print(my_list[i])
    # не забываем увеличивать переменную на единицу    
    i = i + 1
````

````python
# поэлементный перебор списка
for name in ['Винни-Пух', 'Пятачок', 'Сова', 'Осел']:
    print("Привет, Я " + name + '!')
````
